```python
import xarray as xr
import os
import matplotlib.pyplot as plt
```

```python
ncs = [f'GOES_Files/ACMF/{i}' for i in os.listdir('GOES_Files/ACMF/') if i.endswith('.nc')]
sorted(ncs)
```

```python
ds = ncs[0]
xr.open_dataset(ds)
```

```python
# Define el folder donde se guardarán las imágenes
output_folder = 'Images/ACMF/'

# Crea el folder si no existe
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
for nc in ncs:
    ds = xr.open_dataset(nc).sel(x=slice(-0.01, 0.03), y=slice(0.05, 0.01))
    variable = ds['ACM']
    variable.plot()

    # Guardar la figura en el folder especificado como JPG
    output_path = os.path.join(output_folder, f"{os.path.basename(nc).replace('.nc', '.jpg')}")
    plt.savefig(output_path, format='jpg')  # Cambiamos a formato JPG
    
    # Cerrar la figura para liberar memoria
    plt.close()
```
